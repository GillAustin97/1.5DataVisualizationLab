````python
import pandas as pd

df = pd.read_csv("Console_Data.csv")

df[df["Generation"] == 1]["Company"].value_counts().plot(
    kind="bar",
    title="Most Successful Company in the First Generation"
)
````

  
* What company was the most successful in the first year of console wars?
  **Looks like there is a tie for the first year but maybe its cause first gens to try both of them out caused it to be close**

````python
import pandas as pd

df = pd.read_csv("Console_Data.csv")

df["Discontinuation Year"] = pd.to_numeric(df["Discontinuation Year"], errors="coerce")

(df["Discontinuation Year"] - df["Released Year"]).plot(
    kind="hist",
    title="Console Lifespan (Years)"
)
````

* How long does consoles last before being discontinued?
  **^ years seems to be how long consoles last before discontinued**

````python
import pandas as pd

df = pd.read_csv("Console_Data.csv")

df.groupby("Generation")["Units sold (million)"].mean().plot(kind="line",
    title="Average Units Sold by Console Generation"
)
````

* Does newer consoles sell more than older consoles?
  **Newer consoles sell more than the older consoles currently, you can see that the 6-9th gen sells more than the older gens**

````python
import pandas as pd

df = pd.read_csv("Console_Data.csv")

df.plot(
    kind="bar",
    x="Console Name",
    y="Units sold (million)",
    title="Most Popular Console"
)
````

* What is the most popular console this year?
  **Playstation 2 still seems to be the winner!**
